```python
def read_file(exten, file = None, column = None):
    df = None
    if exten == 'csv':
        df = pd.read_csv('data/' + file, encoding = 'utf-8')
    elif exten == 'excel':
        df = pd.read_excel('data/' + file, column, encoding = 'utf-8')
    return df

def export_to_file(exten, df, file_name):
    if exten == 'excel':
        writer = pd.ExcelWriter(file_name)
        df.to_excel(writer,'Sheet1')
        writer.save()
    elif exten == 'csv':
        df.to_csv(file_name, encoding='utf-8', index=False)
```

### Tokenize the arabic synonyms

```python
from itertools import chain

# return list from series of dash-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split('-')))

# calculate lengths of splits
lens = dlm_all[synonym_col].str.split('-').map(len)

dlm_all_expanded = pd.DataFrame({synonym_col: chainer(dlm_all[synonym_col])})
dlm_all_expanded = dlm_all_expanded[synonym_col].str.strip()
```

```python
# drop empty
dlm_all_expanded = dlm_all_expanded.replace('', np.nan, regex=True)
dlm_all_expanded = dlm_all_expanded.dropna()

# drop duplicates
dlm_all_expanded.drop_duplicates(inplace = True)

display(dlm_all_expanded[:122])
```
